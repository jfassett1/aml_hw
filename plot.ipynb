```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu1, sigma1 = 18, 16  # Mean and standard deviation for class 1
mu2, sigma2 = 21, 1.5  # Mean and standard deviation for class 2
pri1, pri2 = 1/3, 2/3  # Prior probabilities

# Define range for x values
x = np.linspace(10, 30, 500)

# Compute probability densities
pdf1 = norm.pdf(x, mu1, sigma1)
pdf2 = norm.pdf(x, mu2, sigma2)

# Find decision boundary where pdf1 = pdf2

# Plot distributions
plt.figure(figsize=(8, 5))
plt.plot(x, pdf1, label='$p(x|\omega_1) P(\omega_1)$', color='blue')
plt.plot(x, pdf2, label='$p(x|\omega_2) P(\omega_2)$', color='red')
# plt.axvline(x=18.6232, color='black', linestyle='--', label='Decision Boundary #1 (18.6232)')
# plt.axvline(x=23.9951, color='black', linestyle='--', label='Decision Boundary #2 (23.9951)')

# plt.axvline(x=19, color='black', linestyle='--', label='$x=19$')

# Labels and legend
plt.xlabel('$x$')
plt.ylabel('Probability Density')
plt.title('Probability Densities of the Classes')
plt.legend()
plt.grid()
plt.show()
```
